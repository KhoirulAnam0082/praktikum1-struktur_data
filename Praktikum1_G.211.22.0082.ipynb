```python
#Deklarasi List
nilai = [1,7,8,5,9,6,14,15,10,2]

#tampilkan list
print("List Nilai : \n",nilai)

#menggunakan count
print("\nJumlah Nilai 5 Dalam List (count) : \n", nilai.count(50))

#menggunakan index
print("\nIndex ke 8 (index) : \n", nilai.index(10))

#menggunakan insert
nilai.insert(1,22)
print("\nTambah Nilai List (Insert) : \n", nilai)

#menggunakan pop
nilai.pop()
nilai.pop()
print("\nList Nilai (pop) :\n", nilai)

#menggunakan remove
nilai.remove(7)
nilai.remove(9)
print("\nList Nilai (remove) : \n", nilai)
```

```python
#menggunakan extend
nilai2 = [4,2,8,30]
nilai.extend(nilai2)
print("\nList Nilai (extend) : \n", nilai)

#menggunakan reverse
nilai.reverse()
print("\nList Nilai (reverse) : \n", nilai)

#menggunakan sort
nilai.sort()
print("\nList Nilai (sort) : \n", nilai)

#menggunakan max
print("\nNilai Terbanyak (max) : \n", max(nilai))

#menggunakan min
print("\nNilai Terendah (min) : \n",(nilai))

#menggunakan sum
print("\nTotal penjumlahan nilai dari list_nilai (sum) : \n", sum(nilai))


```

```python
#Deklarasi List Nama teman
nama_teman = ["ilham","rayhan","tegar","hilmi","nando","alip"]

#Tampilkan List
print("List Nama Orang: \n", nama_teman)

#menggunakan len
print("\nJumlah Index List Nama Orang (len): \n", len(nama_teman))

#menggunakan count
print("\nJumlah Nama tegar dalam list (count): \n", nama_teman.count)

#menggunakan append
print("\nMenambah Nama ke dalam List (append): ")
nama_teman.append("piyan")
nama_teman.append("jono")
print("List Nama Orang: ", nama_teman)

#menggunakan index
print("\nIndex ke-4 (index): \n", nama_teman.index("nando"))

#menggunakan insert
nama_teman.insert(2, "joko")
print("\nMenambah Nama ke dalam List (Insert) : \n", nama_teman)

#menggunakan pop
nama_teman.pop()
nama_teman.pop()
print("\nList Nama Orang (pop): \n", nama_teman)

#menggunakan remove
nama_teman.remove("alip")
nama_teman.remove("rayhan")
print("\nList Nama Orang (remove): \n", nama_teman)
```
